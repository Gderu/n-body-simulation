```python
import matplotlib.pyplot as plt
num_bodies = [2, 3, 4, 4.7, 5]
bf_rayon = [0.00020132064819335937, 0.003598642349243164, 0.32269997596740724, 10.632158398628235 , 42.98040008544922]
bf_python = [0.0006999731063842774, 0.12410016059875488, 10.604949283599854]
plt.plot(num_bodies, bf_rayon, label="rayon")
plt.plot(num_bodies[:3], bf_python, label="python")
plt.xlabel("$log_{10}$ number of bodies")
plt.ylabel("seconds")
plt.legend()
```

-Need to decide way to parallelize the octree construction. One thing I thought up is to have a thread for each node and queues that each node receives of bodies inside of it. This has downsides - lots of threads and memory allocation via the queues.

After think about it, a better approach might be to have a set number of threads and each thread has a node it operates on. Once done with the node, a thread can move on to a different node still in the work. This way we have a set number of threads and denser regions can be parallelized as well. In this case need to find a way to pass task onto child thread when it acquires the node.


-Allocate all memory ahead of time and have maximum depth - costs more memory, but gains in not allocating each time. Look into it

```python
import time
import numpy as np
total = 0
for _ in range(1):
    arr = np.random.randn(100000)
    before = time.time()
    arr.sort()
    total += time.time() - before

print(total / 1)
```
